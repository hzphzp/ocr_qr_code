```python
import PIL
from PIL import ImageFont
from PIL import Image
from PIL import ImageDraw, ImageChops, ImageFilter
import random
import numpy
```

```python
# Definition of some constant
DOC_DIM = (20, 20)
BOARDER_SIZE = 2
FONT_SIZE = 200
FOREGROUND_COLOR = (255, 255, 255)
BACKGROUND_COLOR = (0, 0, 0)
FONT_NAME = "SimSun.ttf"
```

```python
def generate_doc(char, font_name, size):
    IMAGE_SIZE = (size*DOC_DIM[0] + 2*BOARDER_SIZE, size*DOC_DIM[1] + 2*BOARDER_SIZE)
    img=Image.new("RGB", IMAGE_SIZE, FOREGROUND_COLOR)
    font = ImageFont.truetype(font_name, size)
    draw = ImageDraw.Draw(img)
    draw.rectangle([(0, 0), (IMAGE_SIZE[0]-1, IMAGE_SIZE[1]-1)], FOREGROUND_COLOR, BACKGROUND_COLOR)
    try:
        for row in range(DOC_DIM[1]):
            for col in range(DOC_DIM[0]):
                draw.text((BOARDER_SIZE + col*size, BOARDER_SIZE + row*size), char[row*DOC_DIM[0]+col], BACKGROUND_COLOR,font=font)
    except IndexError:
        pass
    return img
```

```python
chars = open('data/frequent_chars.txt', 'r').read()
generate_doc(chars, "fonts/{}".format(FONT_NAME), 40)
```
